```python
import numpy as np
import matplotlib.pyplot as plt

#Create Function
function_x = np.linspace(-3, 3, 100)
function_y = 2 * function_x ** 2 #y = 2x^2

fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(function_x, function_y)

#Start point
x = 3
y = 2 * x ** 2
ax.scatter(x, y, color = "blue", s = 100)

#Set learning rate
lr = 0.02

#Gradient based learning
for _ in range(20):
    dx_dy = 4 * x
    x = x - lr * dx_dy
    y = 2 * x ** 2
    ax.scatter(x, y, color = "red", s = 100)
```
