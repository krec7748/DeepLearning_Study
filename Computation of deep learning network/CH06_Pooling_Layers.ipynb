## 1. Max/Avg Pooling

### 1. Max Pooling

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import MaxPooling1D

#Set Length, pooling size, strides
L, f, s = 12, 2, 1

#Input setting
x = tf.random.normal(shape = (1, L, 1))

#Pooling (Tensorflow)
pool_max = MaxPooling1D(pool_size = f, strides = s)
pooled_max_tf = pool_max(x)

#Pooling (Manual)
x = x.numpy().flatten()
pooled_max_manual = np.zeros(shape = (L - f + 1, ))
for i in range(L - f + 1):
    window = x[i : i + f]
    pooled_max_manual[i] = np.max(window)

#Confirm
print("x : {}\n{} \n".format(x.shape, x))
print("Pooled_max(Tensorflow): {}\n{}\n".format(pooled_max_tf.shape, pooled_max_tf.numpy().flatten()))
print("Pooled_max(Manual): {}\n{}\n".format(pooled_max_manual.shape, pooled_max_manual))
```

### 2. Average Pooling

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import AveragePooling1D

#Set length, pooling size, strides
L, f, s = 12, 2, 1

#Setting input
x = tf.random.normal(shape = (1, L, 1))

#Pooling (Tensorflow)
pool_avg = AveragePooling1D(pool_size = f, strides = s)
pooled_avg_tf = pool_avg(x)

#Pooling (Manual)
x = x.numpy().flatten()
pooled_avg_manual = np.zeros(shape = (L - f + 1, ))
for i in range(L - f + 1):
    window = x[i : i + f]
    pooled_avg_manual[i] = np.mean(window)

#Confirm
print("x : {}\n{} \n".format(x.shape, x))
print("Pooled_max(Tensorflow): {}\n{}\n".format(pooled_avg_tf.shape, pooled_avg_tf.numpy().flatten()))
print("Pooled_max(Manual): {}\n{}\n".format(pooled_avg_manual.shape, pooled_avg_manual))
```

## 2. 2D Max/Avg Pooling

### 1. 2D Max Pooling

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import MaxPooling2D

#Set minibatch, height, width, number of channel
N, n_H, n_W, n_C = 1, 5, 5, 1

#Set pooling size, strides
f, s = 2, 1

#Setting input
x = tf.random.normal(shape = (N, n_H, n_W, n_C))

#Pooling (Tensorflow)
pool_max = MaxPooling2D(pool_size = f, strides = s)
pooled_max_tf = pool_max(x)

#Pooling (Manual)
x = x.numpy().squeeze()
pooled_max_manual = np.zeros(shape = (n_H - f + 1, n_W - f + 1))
for h in range(n_H - f + 1):
    for w in range(n_W - f + 1):
        window = x[h : h + f, w : w + f]
        pooled_max_manual[h, w] = np.max(window)

#Confirm
print("x : {}\n{} \n".format(x.shape, x))
print("Pooled_max(Tensorflow): {}\n{}\n".format(pooled_max_tf.shape, pooled_max_tf.numpy().squeeze()))
print("Pooled_max(Manual): {}\n{}\n".format(pooled_max_manual.shape, pooled_max_manual))
```

## 3. 2D Max/Avg Pooling (Number of channels: 3)

### 1. 2D Max Pooling

```python
from re import S
import math
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import MaxPooling2D

#Set minibatch, height, width, number of channel
N, n_H, n_W, n_C = 1, 5, 5, 3

#Set pooling size, strides
f, s = 2, 2

#Input setting
x = tf.random.normal(shape = (N, n_H, n_W, n_C))

#Pooling (Tensorflow)
pool_max = MaxPooling2D(pool_size = f, strides = s)
pooled_max_tf = pool_max(x)

#Pooling (Manual)
x = x.numpy().squeeze()
pooled_max_manual = np.zeros(shape = (math.floor((n_H - f) / s + 1),
                                      math.floor((n_W - f) / s + 1),
                                      n_C))
for c in range(n_C):
    c_image = x[:, :, c]
    h_idx = 0
    for h in range(0, n_H - f + 1, s):
        w_idx = 0
        for w in range(0, n_W - f + 1, s):
            window = c_image[h : h + f, w : w + f]
            pooled_max_manual[h_idx, w_idx, c] = np.max(window)
            w_idx += 1
        h_idx += 1

#Transpose (for display)
x_t = np.transpose(x, (2, 0, 1))
pooled_max_tf_t = np.transpose(pooled_max_tf.numpy().squeeze(), (2, 0, 1))
pooled_max_manual_t = np.transpose(pooled_max_manual, (2, 0, 1))

#Confirm
print("x : {}\n{} \n".format(x.shape, x_t))
print("Pooled_max(Tensorflow): {}\n{}\n".format(pooled_max_tf.shape, pooled_max_tf_t))
print("Pooled_max(Manual): {}\n{}\n".format(pooled_max_manual.shape, pooled_max_manual_t))
```

## 4. Padding

### 1. ZeroPadding 2D Layer

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import ZeroPadding2D

#Input setting
images = tf.random.normal(shape = (1, 28, 28, 3))

#Padding
zero_padding = ZeroPadding2D(padding = 2)
y = zero_padding(images)

#Confirm
print(images.shape)
print(y.shape)
```

### 2. ZeroPadding with Conv2D Layers

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D

#Input setting
images = tf.random.normal(shape = (1, 28, 28, 3))

#Conv Layer (with padding)
conv = Conv2D(filters = 1, kernel_size = 3, padding = "same") #valid: No padding
y = conv(images)

#Confirm
print(y.shape)
```

## 5. Strides

### 1. Strides in Conv2D Layers

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D

#Input setting
images = tf.random.normal(shape = (1, 28, 28, 3))

#Conv Layer (with strides)
conv = Conv2D(filters = 1, kernel_size = 3, padding = "valid", strides = 2)
y = conv(images)

#Confirm
print(images.shape)
print(y.shape)
```

### 2. Strides in Pooling Layers

```python
import tensorflow as tf
from tensorflow.keras.layers import MaxPooling2D

#Input setting
images = tf.random.normal(shape = (1, 28, 28, 3))

#Pooling Layer (with strides)
conv = MaxPooling2D(pool_size = 3, strides = 2)
y = conv(images)

#Confirm
print(images.shape)
print(y.shape)
```
